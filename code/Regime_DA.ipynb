```python
import h5py
from scipy.io import loadmat

# load data
data_path = './TrueData.mat'
file = loadmat(data_path)
print("Keys: %s" % file.keys())
x = file['x_truth']
y = file['y_truth']
z = file['z_truth']
label = file['S']
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

N = x.shape[1]
data = np.concatenate((x, y, z), axis=0)
data = data.T

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=label, cmap='viridis', s=1)
# cbar = fig.colorbar(scatter, ax=ax)
# cbar.set_label('Labels')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
